```python
import os
from os import path
import h5py

import pybullet as p

import trimesh
import numpy as np
import matplotlib.pyplot as plt
import multiprocessing as mp
```

```python
data_dir = 'data/old/object'
filename_list = os.listdir(data_dir)
def start_process():
    print('Starting', mp.current_process().name)

def modifyObjectData(filename):
    print(filename)
    
    with h5py.File(path.join(data_dir, filename), mode='r') as data:
        mesh_filename = data['filename'][()].decode()
        old_size = data['size'][()]
    new_size = .1 + np.random.RandomState().rand()*.05 # 20 ~ 30 cm     
    
    
    mug_trimesh = trimesh.load(path.join('data/old/meshes', mesh_filename))
    mug_trimesh.apply_scale(new_size/old_size)
    mug_trimesh.export(path.join('data/meshes', mesh_filename))
    
    
    p.vhacd(path.join('data/meshes', mesh_filename), 
            path.join('data/meshes_coll', mesh_filename), 
            "log.txt")

    mug_trimesh = trimesh.load(path.join('data/meshes_coll', mesh_filename))
    com = mug_trimesh.center_mass
    mass = mug_trimesh.mass
    
    with h5py.File(path.join('data/object', filename), mode='w') as data:
        data.create_dataset("filename", data=mesh_filename)
        data.create_dataset("size", data=new_size)
        data.create_dataset("com", data=com)
        data.create_dataset("mass", data=mass)
    
    return mass

pool = mp.Pool(processes=7,
               initializer=start_process)
mass_list = pool.map(modifyObjectData, filename_list)
pool.close() # no more tasks
pool.join()  # wrap up current tasks

plt.plot(mass_list)
plt.show()
```
