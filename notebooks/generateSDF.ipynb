```python
load_dir = 'data/object'
mesh_dir = 'data/meshes_coll'
save_dir = 'data/shape'

def start_process():
    print('Starting', mp.current_process().name)

def saveShapeData(filename):
    print(filename)
    
    data_obj = h5py.File(path.join(load_dir, filename), mode='r')
    size = data_obj['size'][()]
    mesh_name = data_obj['filename'][()]
    data_obj.close()
    
    object_trimesh = trimesh.load(path.join(mesh_dir, mesh_name))
    points, sdf = compute_sdf2(object_trimesh, 
                               N = 5000, 
                               sig = 0.01,
                               scale = size,
                               view = False)
    print(points.shape, sdf.shape)
    
    data = h5py.File(path.join(save_dir, filename), mode='w')
    data.create_dataset("point", data=points)
    data.create_dataset("sdf", data=sdf)
    data.close()
    
    
filename_list = [fn for fn in os.listdir(load_dir) if fn.endswith('.hdf5')]
pool = mp.Pool(processes=32,
               initializer=start_process)
outputs = pool.map(saveShapeData, filename_list)

pool.close() # no more tasks
pool.join()  # wrap up current tasks
```

```python
for filename in filename_list:
    saveShapeData(filename)
```
