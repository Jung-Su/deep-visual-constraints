```python
mug_inds = [28,40,84]
mug_poses = [[-0.3,0.2,1.2,1,0,1,-1.],
             [0.2,0.2,1.2,1,0,0,.0],
             [-0.2,-0.2,1.2,1,0,0,.0]
            ]

S = Simulation("world2.g", mug_inds, mug_poses)
C = Configuration("world2.g", exp_name='PIFO_best', mug_inds=mug_inds)
```

```python
V = ry.ConfigurationViewer()
V.setConfiguration(S.RealWorld)
```

```python
C.setJointState(S.get_q())
out = S.takePicture(mug_inds[:])
C.updatePIFO(mug_inds[:], *out[2:])
Skeleton = [
    ["grasp", "", mug_inds[1]], 
    ["hang", "M_", mug_inds[1]],
    ["grasp", "", mug_inds[0]], 
    ["hang", "U_", mug_inds[0]],
    ["grasp", "", mug_inds[2]], 
    ["hang", "L_", mug_inds[2]]
   ]
traj, komo = C.solveKOMO(Skeleton)
V = komo.view()
```

```python
V = komo.view()
V.playVideo(5.)
```

```python
S = Simulation("world2.g", mug_inds, mug_poses)

S.executeTrajectory(traj[:10])
grasped = S.closeGripper(mug_inds[1])
S.executeTrajectory(traj[10:20])
S.openGripper()
S.stepNone(1.)

S.executeTrajectory(traj[20:30])
grasped = S.closeGripper(mug_inds[0])
S.executeTrajectory(traj[30:40])
S.openGripper()
S.stepNone(1.)

S.executeTrajectory(traj[40:50])
grasped = S.closeGripper(mug_inds[2])
S.executeTrajectory(traj[50:])
S.openGripper()
S.stepNone(1.)

S.goingBack(1.)
S.stepNone(1.)
```

```python
t = 0
save_dir = 'screenshots/multiMug/'

S = Simulation("world2.g", mug_inds, mug_poses)
input()

for i in range(10):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
grasped = S.closeGripper(mug_inds[1])
plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
t += 1
for i in range(10,20):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
S.openGripper()
for i in range(10):
    S.stepNone(0.1)
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
    
    
for i in range(20,30):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
grasped = S.closeGripper(mug_inds[0])
plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
t += 1
for i in range(30,40):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
S.openGripper()
for i in range(10):
    S.stepNone(0.1)
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1

for i in range(40,50):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
grasped = S.closeGripper(mug_inds[2])
plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
t += 1
for i in range(50,60):
    S.executeTrajectory(traj[i:i+1])
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
S.openGripper()
for i in range(10):
    S.stepNone(0.1)
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1

dq = S.qInit - S.get_q()
for i in range(10):
    S.stepPosition(S.get_q()+dq/10, 0.1)
    plt.imsave(save_dir+str(t).zfill(3)+'.png', S.S.getScreenshot()[::-1])
    t += 1
```
